El algoritmo de similitud de nodos compara un conjunto de nodos en funcin de los nodos a los que estn conectados. Los algoritmos de similitud calculan la similitud de pares de nodos utilizando diferentes mtricas basadas en vectores.

## Distancia euclidia

## Que es ?

es un algoritmo de similaridad que aplica una formula matemtica que nos permite conocer la distancia en linea recta que separa a dos puntos ubicados en un espacio n-dimensional. Gracias a ese calculo podemos determinar un indice de similitud que puede formar parte fundamental de sistemas de recomendacin

## usos

La herramienta Distancia euclidiana se utiliza con frecuencia como una herramienta independiente para algunas aplicaciones, como la bsqueda del hospital ms cercano para un vuelo de helicptero de emergencia. Alternativamente, esta herramienta se puede utilizar para crear un mapa de adecuacin, donde se necesitan datos que representen la distancia desde un objeto determinado.

Ejercicio en Python 


```python
from neo4j import GraphDatabase

uri = "bolt://localhost:7687"
driver = GraphDatabase.driver(uri,auth=("neo4j", "p4t1t01234"),encrypted=False )


def get_friends_of(tx, name):
    friends = []
        
    result = tx.run("MATCH (p1:Ciudad {name: $name})-[likes1:LIKES]->(Ciudad)"
                   "MATCH (p2:UPC {})-[likes2:LIKES]->(UPC)"
                   "RETURN p1.name AS PAIS,"
                   "p2.name AS CIUDAD,"
                   "gds.alpha.similarity.euclideanDistance(collect(likes1.score), collect(likes2.score)) AS similarity", name=name)

    for record in result:
        friends.append(record['CIUDAD'])
        friends.append(record['similarity'])
    return friends

with driver.session() as session:
    friends = session.read_transaction(get_friends_of, 'Manta')
    for friend in friends:
        print(friend)

driver.close()
```
