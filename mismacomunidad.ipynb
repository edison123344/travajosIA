# Algoritmos de deteccin de la comunidad
Los algoritmos de deteccin de comunidad se utilizan para evaluar cmo se agrupan o particionan los grupos de nodos, as como su tendencia a fortalecerse o separarse.
Esto significa qu tan densamente conectados estn los nodos dentro de una comunidad, en comparacin con lo conectados que estaran en una red aleatoria.
## usos
La estructura de comunidades es bastante comn en redes reales. Las redes sociales incluyen grupos comunitarios basados en ubicaciones comunes, intereses comunes, ocupaciones comunes, etc. Las redes metablicas tienen comunidades basadas en grupos funcionales. En las redes de citaciones se forman comunidades por temas de investigacin.Ser capaz de identificar estas sub-estructuras dentro de una red puede ayudar a entender de forma intuitiva cmo el funcionamiento de la red y su topologa se afectan el uno al otro. Tal intuicin puede ser til en mejorar algunos algoritmos sobre grafos tales como agrupamiento espectral
## Ejemplo en NEO4J

CALL gds.alpha.closeness.stream({  
  nodeProjection:         'Node',  
  relationshipProjection: 'LINK'  
})  
YIELD nodeId, centrality  
RETURN gds.util.asNode(nodeId).id AS user, centrality  
  ORDER BY centrality DESC  
* ver si pertenese ala misma comunidad







```python
from neo4j import GraphDatabase

uri = "bolt://localhost:7687"
driver = GraphDatabase.driver(uri,auth=("neo4j", "p4t1t01234"),encrypted=False )


def get_friends_of(tx):
    friends = []
        
    result = tx.run("MATCH (p1:Colegio {})"
                    "MATCH (p2:Colegio {})"
                    "RETURN gds.alpha.linkprediction.sameCommunity(p1, p2) AS Puntaje, p1.name, p2.name")
    
    
    
    for record in result:
        friends.append(record['Puntaje'])
        friends.append(record['p1.name'])
        friends.append(record['p2.name'])
    return friends

with driver.session() as session:
    friends = session.read_transaction(get_friends_of)
    for friend in friends:
        print(friend)

driver.close()
```
