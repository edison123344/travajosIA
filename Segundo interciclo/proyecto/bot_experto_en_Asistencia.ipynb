## prueva de el funcionamiento del chatbot 

```python
from chatterbot import ChatBot
from chatterbot.trainers import ListTrainer

# Create a new chat bot named Charlie
chatbot = ChatBot('Charlie')

trainer = ListTrainer(chatbot)

trainer.train([
    "Hi, can I help you?",
    "Sure, I'd like to book a flight to Iceland.",
    "Your flight has been booked."
])

# Get a response to the input text 'I would like to book a flight.'
response = chatbot.get_response('help you?')

print(response)
```

## asistente de chatbot  con mongodb

```python
bot = ChatBot(
    'Asistente ',
    # Adaptador de almacenamiento
    storage_adapter='chatterbot.storage.MongoDatabaseAdapter',
    database_uri='mongodb://localhost:27017/chatbotAsistente2',
    #adaptador de txt
    input_adapter= "chatterbot.input.TerminalAdapter",
    output_adapter= "chatterbot.output.OutputAdapter",
    output_format= "text",
   
    logic_adapters=[
        {
            "import_path": "chatterbot.logic.BestMatch",
            "statement_comparison_function": "chatterbot.comparisons.levenshtein_distance",
            "response_selection_method": "chatterbot.response_selection.get_most_frequent_response"
        },
        {# Adaptador para respuestas especificas
            'import_path': 'chatterbot.logic.SpecificResponseAdapter',
            'input_text': 'Quiero hacer una consulta?',
            'output_text': 'cual es tu dilema'
        },
    ],
    
    preprocessors=[
        'chatterbot.preprocessors.clean_whitespace'
    ],#Adaptador de apredisaje
     read_only=False,
)
print('asistente chatbot')

# Create a new chat bot named Charlie
chatbot = ChatBot('Charlie')

trainer = ListTrainer(chatbot)


# Get a response to the input text 'I would like to book a flight.'

while True:
    try:
        user_input = input()
        
        bot_response = bot.get_response(user_input)

        print(bot_response)

    # Press ctrl-c or ctrl-d on the keyboard to exit
    except (KeyboardInterrupt, EOFError, SystemExit):
        break
```
